```python
import matplotlib.pyplot as plt 

import numpy as np 

import numpy as np
import matplotlib.pyplot as plt

categories = ['A','B','C','D','E']
values = [38,67,45,34,63]

x = np.arange(len(categories))

fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(
    x, values, width=0.5, color='skyblue',
    edgecolor='black', hatch='/', alpha=0.8
)

ax.set_title('Bar Plot Design Mastery by Nikunj', fontsize=20, fontweight='bold')
ax.set_xlabel('Categories', fontsize=15)
ax.set_ylabel("Values", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12)

ax.grid(axis='y', linestyle='--', alpha=0.8)

plt.show()


#       HORIZONTAL BAR EXAMPLE 
fig2, ax2 = plt.subplots(figsize=(10, 6))
bars_h = ax2.barh(
    categories, values, 
    color='lightgreen', 
    edgecolor='black', 
    hatch='o', 
    alpha=0.85
)
ax2.set_title("Horizontal Bar Design", fontsize=18, fontweight='bold')
ax2.set_xlabel("Values", fontsize=14)
ax2.bar_label(bars_h, fmt='%.1f', padding=3, fontsize=12)
ax2.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
