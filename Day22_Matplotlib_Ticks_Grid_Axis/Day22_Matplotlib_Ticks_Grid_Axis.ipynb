## 1. plt.xticks() and plt.yticks()

```python
# Ye functions tick positions aur labels define karte hain.

# we can specify :-
#Kahan tick dikhega
#Uska label kya ho
#Label rotate karna hai ya nahi
```

```python
import matplotlib.pyplot as plt 
plt.xticks([1,2,3], ['one','two','three'], rotation = 45)
plt.yticks([10,20.30,40])
```

# 2. Adding & Styling Gridlines with plt.grid()

```python
plt.grid(True , color ='gray',linestyle ='--', linewidth = 0.5)
```

## 3. Adjusting Axis Range using plt.xlim() and plt.ylim()

```python
plt.xlim(0.5, 5.5)
plt.ylim(0, 60)
```

## Full Example

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.plot(x, y, marker='o')

#  Custom tick labels on x-axis with rotation
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'], rotation=45)

# Custom tick positions on y-axis
plt.yticks([10, 20, 30, 40, 50])

# Add gridlines
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)

# Set axis limits
plt.xlim(0.5, 5.5)
plt.ylim(0, 60)

# Titles and Labels
plt.title("Customized Ticks and Grid")
plt.xlabel("Custom Categories")
plt.ylabel("Values")

#  Prevent label overlap
plt.tight_layout()

plt.show()
```
