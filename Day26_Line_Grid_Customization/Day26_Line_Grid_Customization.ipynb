## 1. Custom Line Styles

```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]

plt.plot(x,y, color ='green',linestyle = '--' , linewidth = 2)
plt.title('dashed line ')
plt.show()
```

## 2. Adding Gridlines

```python
plt.plot(x,y , color ='cyan',linestyle =':')
plt.grid()
plt.title("With grid this time YO ")
plt.show()
```

## 3. Complete Example with Grid and Line Customization

```python
plt.plot(x,y , color ='yellow',linestyle ='-.', linewidth =2 )
plt.title("Customized line with grid ")
plt.xlabel('X-Axix')
plt.ylabel('Y-label')
plt.grid(True, color = 'green', linestyle = '--', linewidth = 1.5,alpha = 1)
plt.show()
```
