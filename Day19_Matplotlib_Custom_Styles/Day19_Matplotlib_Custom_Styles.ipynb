## 1. Setting Colors and Line Styles

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y, color = 'red', linestyle = '--',linewidth = 2 )
plt.title("Line Plot with Custom Styles")
```

## 2. Customizing Markers

```python
plt.plot(x,y,marker ='o' , markersize = 8 , markerfacecolor = 'yellow' , markeredgewidth = 2)
```

##  3. Figure Size and DPI

```python
plt.figure(figsize = (10,6),dpi = 100)
```

## 4. Adding Grid

```python
plt.grid(True, linestyle = '--', color ='gray' , alpha = 0.7)
```

##  5. Background Color

```python
fig = plt.figure()
fig.patch.set_facecolor('lightblue')
```
