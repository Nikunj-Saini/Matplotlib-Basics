## 1. Line Labels and Legends

### When you have multiple lines on one plot, Then we use label and  plt.legend() to show which line represents what.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5] 
student_A_scores = [2, 3, 5, 7, 11]
student_B_scores = [1, 4, 6, 8, 10]
```

```python
plt.plot(x,student_A_scores, label ='student A', color = 'blue', marker = 'o')
plt.plot(x,student_B_scores, label ='student B', color = 'green', marker = 's')
plt.xlabel('Exam Number')
plt.ylabel('Score')
plt.title('Exam Scores of Two Students')
plt.legend()
```

## 2. Annotations

### Sometimes we want to point out a special value, like the highest score or a turning point then we use the plt.annotate() function .

```python
plt.annotate('Highest Score',
             xy=(5, 11),               # Arrow tip at (x=5, y=11)
             xytext=(3.5, 12),         # Text slightly above
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```
