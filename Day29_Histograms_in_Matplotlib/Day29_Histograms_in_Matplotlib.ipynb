```python
import matplotlib.pyplot as plt
```

```python
import numpy as np 
```

Sample Data

```python
data = np.random.randn(1000)
```

Basic Histogram

```python
plt.hist(data,bins = 20)
```

 Customizations

```python
plt.title("Histogram of Random Data", fontsize=16, fontweight='bold')
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

 Multiple histograms

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # shifted mean

plt.hist(data1, bins=20, color='skyblue', alpha=0.7, label='Data 1')
plt.hist(data2, bins=20, color='salmon', alpha=0.7, label='Data 2')

plt.title("Overlapping Histograms", fontsize=16, fontweight='bold')
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
data = np.random.randn(1000)  

plt.hist(data, bins=20, color='skyblue', edgecolor='black')

plt.title("Histogram of Random Data", fontsize=16, fontweight='bold')
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # shifted mean

plt.hist(data1, bins=20, color='skyblue', alpha=0.7, label='Data 1')
plt.hist(data2, bins=20, color='salmon', alpha=0.7, label='Data 2')

plt.title("Overlapping Histograms", fontsize=16, fontweight='bold')
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
