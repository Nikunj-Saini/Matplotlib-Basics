```python
import matplotlib.pyplot as plt
```

 Pie charts are great for showing proportional data â€” How a whole is divided into parts.

## 1: Define the data

```python
labels = ['Python','Java','C++','Javascipt']
sizes = [35,30,20,15]
colors = ['pink','yellow','cyan','lightgreen']
explode = (0.1,0,0,0)
```

## 2: Create the pie chart

```python
plt.figure(figsize = (6,6))
plt.pie(sizes, labels = labels,colors = colors, explode = explode,shadow = True)
```

## 3: Finalize and show

```python
languages = ['Python', 'Java', 'C++', 'JavaScript']
popularity = [40, 30, 20, 10]

plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title('Programming Language Popularity')
plt.axis('equal')
plt.show()
```
