## 1)  Importing the Liberary

```python
import matplotlib.pyplot as plt
```

 ## 2) Basic Line Plot:

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.show()
```

## 3) Adding labels and title:

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X Axix")
plt.ylabel("Y Axis")
plt.show()
```

## 4) Customizing Lines:

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
```
