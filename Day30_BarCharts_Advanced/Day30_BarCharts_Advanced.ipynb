```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np 
```

```python
# Data
categories = ['A', 'B', 'C', 'D']
values1 = [20, 34, 30, 35]
values2 = [25, 32, 34, 20]

# Positions
x = np.arange(len(categories))
width = 0.35  

# Create plot
fig, ax = plt.subplots(figsize=(8, 6))

# Bars for dataset 1
bars1 = ax.bar(x - width/2, values1, width, label='Dataset 1',
               color='skyblue', edgecolor='black', hatch='/', alpha=0.9)

# Bars for dataset 2
bars2 = ax.bar(x + width/2, values2, width, label='Dataset 2',
               color='salmon', edgecolor='black', hatch='\\', alpha=0.9)

# Add labels and title
ax.set_ylabel('Scores')
ax.set_xlabel('Categories')
ax.set_title('Grouped Bar Chart with Advanced Styling')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Annotate values on bars
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 0.5,
            f'{height}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
