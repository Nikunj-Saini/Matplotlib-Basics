```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np 
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
product_A = [200, 240, 300, 280, 350, 400]
product_B = [180, 210, 260, 220, 300, 330]
product_C = [150, 190, 220, 200, 270, 290]
profit_margin = [15, 18, 20, 19, 21, 25]

x = np.arange(len(months))
width = 0.25 
```

## 1 Grouped Bar Chart

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width, product_A, width, label="Product A", color="skyblue")
ax.bar(x, product_B, width, label="Product B", color="lightgreen")
ax.bar(x + width, product_C, width, label="Product C", color="salmon")

ax.set_title('Monthly Sales of Products',fontsize = 20,fontweight = 'bold')
ax.set_xticks(x)
ax.set_xticklabels(months)
ax.set_yticklabels("sales (in units")
ax.legend()
ax.grid(axis = 'y', linestyle = "--",alpha = 0.7)
plt.show()
```

# 2. Stacked Bar Chart

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(months, product_A, label="Product A", color="skyblue")
ax.bar(months, product_B, bottom=product_A, label="Product B", color="lightgreen")
ax.bar(months, product_C, bottom=np.array(product_A) + np.array(product_B),
       label="Product C", color="salmon")

ax.set_title("Total Monthly Sales (Stacked)", fontsize=16, fontweight="bold")
ax.set_ylabel("Sales (in units)")
ax.legend()
plt.show()
```

# 3. Line Chart with Profit Margin

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(months, product_A, marker="o", label="Product A")
ax.plot(months, product_B, marker="s", label="Product B")
ax.plot(months, product_C, marker="^", label="Product C")
ax.plot(months, profit_margin, marker="*", linestyle="--", label="Profit Margin (%)")

ax.set_title("Sales & Profit Margin Trend", fontsize=16, fontweight="bold")
ax.set_ylabel("Values")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

# 4. Pie Chart (Total Sales Contribution)

```python
total_sales = [sum(product_A), sum(product_B), sum(product_C)]
labels = ["Product A", "Product B", "Product C"]
colors = ["skyblue", "lightgreen", "salmon"]

fig, ax = plt.subplots(figsize=(8, 6))
ax.pie(total_sales, labels=labels, autopct="%.1f%%", colors=colors,
       startangle=90, wedgeprops={"edgecolor": "black"})
ax.set_title("Total Sales Contribution", fontsize=16, fontweight="bold")
plt.show()
```

# 5. Scatter Plot (Sales vs Profit Margin)

```python
avg_sales = [(a+b+c)/3 for a, b, c in zip(product_A, product_B, product_C)]

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(avg_sales, profit_margin, s=100, c="purple", alpha=0.7, edgecolors="black")

ax.set_title("Average Sales vs Profit Margin", fontsize=16, fontweight="bold")
ax.set_xlabel("Average Sales")
ax.set_ylabel("Profit Margin (%)")

for i, month in enumerate(months):
    ax.text(avg_sales[i]+2, profit_margin[i]+0.5, month)

plt.show()
```
