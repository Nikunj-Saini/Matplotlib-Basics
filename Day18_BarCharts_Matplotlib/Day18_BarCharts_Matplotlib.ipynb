## 1. Basic Bar Chart

```python
import matplotlib.pyplot as plt
```

```python
categories = ['MATHMATICS', 'ENGLISH','SCIENCE','HINDI','SOCIAL SCIENCE']
scores = [75,89,79,80,77]

plt.bar(categories, scores)
plt.title("Student Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

##  2. Customizing Bar Charts

```python
colors = [ 'red', 'pink' ,'orange','yellow']
plt.bar(categories, scores, color=colors, width=0.5)
plt.title("Student Scores")
plt.grid( axis ='y')
plt.show()
```

## 3. Horizontal Bar Chart

```python
plt.barh(categories, scores, color = 'skyblue')
plt.title("Scores per student")
plt.xlabel("scores")
plt.ylabel("Subjects")
plt.show()
```

## 4. Adding Value Labels to Bars

```python
bars = plt.bar(categories, scores)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(height), ha='center')
plt.title("Scores of the student")
plt.show()
```
