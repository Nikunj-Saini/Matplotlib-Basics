## STACKED BAR CHART 

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values1 = [30, 40, 20, 50]
values2 = [20, 30, 25, 15]

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(categories, values1, label='Group 1', color='skyblue', edgecolor='yellow')
ax.bar(categories, values2, bottom=values1, label='Group 2', color='salmon', edgecolor='black')
ax.set_title('Stacked Bar Plot', fontsize=20, fontweight='bold')
ax.legend()

plt.show()
```

## PIE CHART

```python
sizes = [34,64,64,23]
colors = ['pink','orange','cyan','green']
explode = (0.05,0.05,0,0)
plt.figure(figsize=(6,6))
plt.pie(sizes,labels = categories, colors = colors,explode = explode ,shadow = True  )
plt.show()
```

##  SCATTER PLOT 

```python
x= np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 500 *np.random.rand(50)
plt.figure(figsize = (8,6))
plt.scatter(x,y,s = sizes , c= colors , alpha = 0.6 ,edgecolor = 'black')
plt.title("Scatter Plot Example", fontsize=16, fontweight='bold')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label='Color Scale')
plt.show()
```
