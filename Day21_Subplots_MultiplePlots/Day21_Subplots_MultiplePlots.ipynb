## Why to use Subplots?

### Sometimes, we want to show multiple charts in one figure for easy comparison or compact layout. Matplotlib offers two main ways to do this:

## 1. plt.subplot() - It is a old style , simple subplots.

```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4]
y1 = [10,20,25,30]
y2 = [30,35,40,45]
```

```python
plt.subplot(2,1,2)
plt.plot(x,y1)
plt.title("First plot")

plt.subplot(2,1,2)
plt.plot(x,y2,color ='green')
plt.title("Second plot")

plt.tight_layout()
plt.show()
```

## 2. plt.subplots() â€” It is more Modern and More Powerful

```python
fig, axes = plt.subplots(2,2)

x = [1,2,3]
y = [2,4,6]

for ax in axes.flat:
    ax.plot(x,y)
    plt.tight_layout()
    
    ax.set_title("Simple line")
```
