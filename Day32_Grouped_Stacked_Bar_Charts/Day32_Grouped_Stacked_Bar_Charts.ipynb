```python
import numpy as np 
import matplotlib.pyplot as plt 

categories= ['A','B','C','D','E']
values1 = [34,52,85,52,38]
values2 = [27,73,58,73,58]

x = np.arange(len(categories))
width = 0.35 

fig1, ax1 = plt.subplots(figsize=(10, 6))
bars1 = ax1.bar(x - width/2, values1, width, label='Dataset 1', 
                color='skyblue', edgecolor='black', hatch='/', alpha=0.9)
bars2 = ax1.bar(x + width/2, values2, width, label='Dataset 2', 
                color='orange', edgecolor='black', hatch='\\', alpha=0.9)

ax1.set_title("Grouped Bar Chart", fontsize=18, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(categories, fontsize=12)
ax1.set_ylabel("Values", fontsize=14)
ax1.legend()
ax1.grid(axis='y', linestyle='--', alpha=0.7)


ax1.bar_label(bars1, padding=3, fontsize=10)
ax1.bar_label(bars2, padding=3, fontsize=10)

plt.show()


fig2, ax2 = plt.subplots(figsize=(10, 6))
bars_stack1 = ax2.bar(categories, values1, label='Dataset 1',
                      color='lightgreen', edgecolor='black', hatch='o', alpha=0.85)
bars_stack2 = ax2.bar(categories, values2, bottom=values1, label='Dataset 2',
                      color='salmon', edgecolor='black', hatch='x', alpha=0.85)


ax2.set_title("Stacked Bar Chart", fontsize=18, fontweight='bold')
ax2.set_ylabel("Values", fontsize=14)
ax2.legend()
ax2.grid(axis='y', linestyle='--', alpha=0.7)

ax2.bar_label(bars_stack1, padding=3, fontsize=10)
ax2.bar_label(bars_stack2, padding=3, fontsize=10)

plt.show()
```
